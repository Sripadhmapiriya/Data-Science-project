1. Make a three-dimensional plot with randomly generate 50 data points for x, y, and z. Set
the point color as red, and size of the point as 50

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate 50 random data points for x, y, and z
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Set point color as red and size as 50
ax.scatter(x, y, z, c='red', s=50)

# Show the plot
plt.show()
```

2.Perform Univariate analysis with the following pandas DataFrame
'points': [1, 1, 2, 3.5, 4, 4, 4, 5, 5, 6.5, 7, 7.4, 8, 13, 14.2]
'assists': [5, 7, 7, 9, 12, 9, 9, 4, 6, 8, 8, 9, 3, 2, 6]
'rebounds': [11, 8, 10, 6, 6, 5, 9, 12, 6, 6, 7, 8, 7, 9, 15]

```python
import pandas as pd

# Create the DataFrame
data = {'points': [1, 1, 2, 3.5, 4, 4, 4, 5, 5, 6.5, 7, 7.4, 8, 13, 14.2],
        'assists': [5, 7, 7, 9, 12, 9, 9, 4, 6, 8, 8, 9, 3, 2, 6],
        'rebounds': [11, 8, 10, 6, 6, 5, 9, 12, 6, 6, 7, 8, 7, 9, 15]}

df = pd.DataFrame(data)

# Descriptive statistics for each variable
points_stats = df['points'].describe()
assists_stats = df['assists'].describe()
rebounds_stats = df['rebounds'].describe()

print("Points statistics:\n", points_stats)
print("\nAssists statistics:\n", assists_stats)
print("\nRebounds statistics:\n", rebounds_stats)
```

This will output summary statistics for each variable, including count, mean, std (standard deviation), min, 25%, 50%, 75%, and max. You can further analyze and visualize the data based on these statistics.

3.i. Using various functions in numpy library, mathematically calculate the values for a
normal distribution and create Histograms to plot the probability distribution curve
ii. Using plt.contour(), plt.contourf(), plt.imshow(), plt.colorbar(), plt.clabel() functions
visualize a contour plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data for normal distribution
mu, sigma = np.mean(df['points']), np.std(df['points'])
data_points = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

# Calculate the probability density function (PDF) for the normal distribution
pdf_values = norm.pdf(data_points, mu, sigma)

# Create histogram
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(df['points'], bins=10, density=True, alpha=0.7, color='blue')
plt.plot(data_points, pdf_values, 'r-', label='Normal Distribution')
plt.title('Histogram and Normal Distribution')
plt.xlabel('Points')
plt.ylabel('Probability')
plt.legend()

# Visualize contour plot
plt.subplot(1, 2, 2)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2 - Y**2)

contour_plot = plt.contour(X, Y, Z, cmap='viridis')
plt.colorbar(contour_plot, label='Contour Values')
plt.title('Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

5.Apply and explore various plotting functions on UCI data set for performing the following:
d) Correlation and scatter plots
e) Histograms
f) Three-dimensional plotting

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Load your UCI dataset (replace 'your_dataset.csv' with your actual file)
data1 = "/content/gdrive/MyDrive/diabetes.csv"
data= pd.read_csv(data1)

# d) Correlation and scatter
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')

# Scatter plot
data.plot(kind='scatter',x='BloodPressure',y='SkinThickness')
plt.show()

# e) Histograms
data.hist(figsize=(10, 8), bins=20)
plt.suptitle('Histograms for Variables', y=1.02)
plt.show()

# f) Three-dimensional plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Pregnancies'], data['Glucose'], data['Insulin'], c=data['BloodPressure'], marker='o', s=50, cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Three-dimensional Plot')

plt.show()
```
